```python
import pandas as pd
import numpy as np
import re
```

```python
%%time
date = pd.read_hdf("../hdf/unreduced_train_date.hdf")
```

```python
%%time
categorical = pd.read_hdf("../hdf/unreduced_train_categorical.hdf")
```

```python
%%time
numeric = pd.read_hdf("../hdf/train_numeric.hdf")
```

```python
columns = [c.replace("F", "C") for c in categorical.columns]
columns = np.concatenate([date.columns, numeric.columns, columns])
columns = [
    (
        int(re.sub(r"^L[0-9]*_S[0-9]*_[FDC]", "", c)),
        c
    )
    for c in columns
]
```
